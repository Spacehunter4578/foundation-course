Get the standard deviation of the values in mat

```python
z = np.array([[100, 90, 135],
                [5, 60, 25],
                [20, 10, 60]])

std_deviation = np.std(z)

print(std_deviation)
```

Get the sum of all the columns in mat

```python
mat = np.array([[12, 13, 14, 15],
                [17, 18, 19, 20],
                [22, 23, 24, 25]])

column_sums = np.sum(mat, axis=0)

print(column_sums)
```
